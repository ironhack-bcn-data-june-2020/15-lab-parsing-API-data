In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

```python
# your code here
#Import Libraries
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
from time import sleep
# headers definition

```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows Phone 10.0; Android 6.0.1; Microsoft; RM-1152)'}

url=f'https://www.zalando.fr/api/catalog/articles?categories=promo-homme&limit=84&offset=0&sort=popularity'
response = requests.get(url, headers=headers).json()
response
num_pages = response['pagination']['page_count']
print(num_pages)
# Your code
articles = pd.DataFrame(response['articles'])
articles
```

```python
num_pages = 10
for offset in range(1,num_pages):
    url=f'https://www.zalando.fr/api/catalog/articles?categories=promo-homme&limit=84&offset={offset*84}&sort=popularity'
    response = requests.get(url, headers=headers).json()
    artic = pd.DataFrame(response['articles'])
    artic.set_index('sku')
    articles = articles.append(artic)
```

## Higher discount

```python
def extract_discount(x):
    for l in x[0]:
        if x[0][l] == 'discountRate':
            return x[0]['value'][x[0]['value'].find('-'):x[0]['value'].find('%')]
    return 0
articles['discount'] = articles['flags'].apply(extract_discount).astype(str).astype(int)
articles.nsmallest(1,'discount')
```
