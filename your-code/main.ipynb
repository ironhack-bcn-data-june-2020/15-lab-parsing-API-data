```python
def get_page_info():
    
    url='https://www.zalando.fr/api/catalog/articles'
    params = {'categories':'promo-enfant', 'limit':84, 'offset':0, 'sort':'popularity'}
    headers = {"User-agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/605.1.15 (KHTML, like Gecko)", "Accept-Encoding":"Application/json"}
    request = requests.get(url, headers=headers, params=params)
    
    fp = urllib.request.urlopen(url)
    mybytes = fp.read()
    response = mybytes.decode("ISO-8859-1")
    fp.close()
    
    return json.loads(response)
```
